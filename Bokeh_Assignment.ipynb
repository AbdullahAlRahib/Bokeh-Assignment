# **Que1**

**How can you create a Bokeh plot using Python code?**

**Answer:**




```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
output_notebook()
p = figure()
p.title.text = "Scatter Plot"
p.xaxis.axis_label = "X-axis"
p.yaxis.axis_label = "Y-axis"
p.grid.grid_line_alpha = 0.5
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=10, color="red")
show(p)
```

# **Que3**

**How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?**

    
**Answer:**

Axes customization:

 * Title: Set the title of the plot using figure.title.text.
 * Axis labels: Set the x-axis and y-axis labels using figure.xaxis.axis_label and figure.yaxis.axis_label, respectively.
 * Axis range: Customize the range of values displayed on the x-axis and y-axis using figure.x_range and figure.y_range.
 * Tick labels and tick formatting: Control the appearance and formatting of tick labels using figure.xaxis[0].formatter and figure.yaxis[0].formatter.
 * Grid lines: Show or hide grid lines on the plot using figure.xgrid.visible and figure.ygrid.visible.

Legend customization:

 * Title: Set the title of the legend using figure.legend.title.
 * Position: Control the position of the legend using figure.legend.location (e.g., "top_left", "bottom_right").
 * Orientation: Set the orientation of the legend using figure.legend.orientation (e.g., "vertical", "horizontal").
 * Label customization: Customize the appearance of legend labels using figure.legend.label_text_font_size, figure.legend.label_text_color, etc.

```python
p = figure()

# Customize the appearance
p.title.text = "My Bokeh Plot"
p.title.text_font_size = "18pt"
p.xaxis.axis_label = "X-axis"
p.yaxis.axis_label = "Y-axis"
p.legend.title = "Legend Title"
p.legend.location = "top_left"
p.legend.label_text_font_style = "italic"

# Add glyphs to the plot
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=10, color="red", alpha=0.5, legend_label="Data")

# Display the plot
show(p)
```

# **Que4**

**What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?**

**Answer:**

A Bokeh server is a Python server application that allows you to create and deploy interactive plots with real-time updates. With Bokeh server, you can build web applications that serve dynamic visualizations, enabling users to interact with and modify the plots directly in a web browser.


The Bokeh server handles communication between the web browser and the Python server, allowing for synchronized updates and interactivity. By using Bokeh server, you can create powerful and responsive data-driven applications where changes made by users or changes in the underlying data can be reflected in the plots in real time, providing a highly interactive and engaging experience.
